## User Authentication

### Password Authentication

Basic idea is that the user has a secret password and user checks password to authenticate the user.<br>

Hashing the password with a function. When the user enters password, the system computes hash(password) and checks if the user entered correct password.<br>

Unix password system has 25xDES hash function. Salting the password makes it harder to conduct a dictionary attack.<br>

* Salt: a random data that is used as an additional input to password hash function.

        Advantage:
        * One password hashed 2^12 ways.


*Playback* attack: when an attacker records your packet and sends it back to server. This attack type still works even if the password is encrypted.<br>

AP4.0 does not protect against server database reading.<br>
AP5.0 uses nonce and public key cryptography to mitigate attacks.<br>

### Biometrics

Using a person's physical characteristics e.g. fingerprint, voice etc. to authenticate. Usually a combination of biometrics and pin or token or multiple biometrics is used to enhance security.<br>

### Trusted Intermediaries

* Trusted Key Distribution Center (KDC):
        
        server shares different secret keys with each user
        users only know their own symmetric keys to communicate with KDC
        users communicate using R1 as session key for shared symmetric encryption.

* Trusted Certification Authority (CA):

        binds public key to certain entity, E.
        digitally signs the public key of E after authentication.
        when users want to communicate, they are given the certificate and CA public key
        they apply the public key to other user's certificate to get their public key
        
**Problems**:<br>

* Single point of failure
* Single trusted entity by all participants
<br>

However, these can be solved by breaking intermediaries into multiple domains.<br>
