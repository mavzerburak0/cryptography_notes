## Wireless Security

Cabir worm can infect cell phones with Symbian OS (outdated).

**Base Station Approach**<br>

Wireless host communicates with a base station (AP).

**Ad Hoc Network Approach**<br>

1. No AP
2. Wireless hosts communicate with each other

**CSMA (Carrier Sense Multiple Access)**<br>

One connection at a time.<br>
Collisions might still occur e.g. if two connections do not hear each other's transmission<br>

**CSMA/CD (Collision Detection)**<br>

This helps detect collisions that might occur during CSMA. It is rather more difficult in Wireless LANs than in Wired LANs where you can measure signal strengths, compare what's transmitted etc.<br>

**Collision Avoidance Mechanisms** solve the problem of two nodes, that do not hear about each other's transmission and wasting bandwidth by creating small reservation packets and make the nodes track the reservation intervals with NAV (network allocation vector).<br>

**RTS-CTS Exchange**: sender transmits a Request-to-Send and receiver replies with Clear-to-Send message. A NAV is established and hidden nodes do not send anything during the specified NAV.<br>

**IEEE 802.11** allows for:
1. CSMA
2. CSMA/CA 
3. polling from AP
<br>

**802.11b** has the following built-in security features:
1. Service Set Identifier (SSID)
2. Differentiation of one AP from another
3. SSID casts "beacon frames" (in plain text format) every few seconds.
<br>

**AP Association** can happen in two ways: **shared key** or **open key** authentication. Open keys allows anyone to start communication with AP whereas shared key adds one extra layer of security by requiring authentication.<br>

### WEP (Wired Equivalent Protocol)

WEP uses 40bit Rivest Cipher (RC4) for encryption with the intention of making wireless connections as secure as wired ones. *RC4 has been found vulnerable.*<br>

**Netstumbler** and **Airsnort** can be used to attack networks.<br>

1. Attacker sets the NIC drivers to Monitor Mode
2. Begins capturing packets with Airsnort
3. Airsnort determines SSID
4. Sessions are saved in Airsnort and a few 1.5 hour sessions will yield the encryption key (*https://www.systutorials.com/docs/linux/man/1-airsnort/*)
<br>

#### More Attacks on Wireless Networks

* Rogue Access Points
* Radio Frequency (RF) Interference
* AP Impersonation
* Man-in-the-Middle Attack
* DDoS Attack
<br>

Lastly, the latest trend in wireless networks is IEEE 802.16 WirelessMAN Standard for its **comprehensive and modern security (with DES and AES)**, **use of key management protocols (RSA)** and **broad bandwidth**.
